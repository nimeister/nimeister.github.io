
## Problem 1 — Array creation & spacing (10 pts)

1. Create an array `a` using `np.arange` that starts at `-2.5`, ends **before** `3.0`, with a step of `0.5`.
2. Create an array `b` using `np.linspace` from `-2.5` to `3.0` (inclusive) with **12** evenly spaced values.
3. In **one sentence**, explain why `a` and `b` have different lengths.


```python

a = np.arange(-2.5, 3.0, 0.5)
b = np.linspace(-2.5, 3.0, num=12, endpoint=True)
explain_a_b = """
`np.arange` is step-based and excludes the stop value (3.0), while `np.linspace`
returns a specified count and, with `endpoint=True`, includes both endpoints — hence different lengths.
"""
a, b, explain_a_b
```

```python
Tjul = T['JUL']
print(type(Tjul))
CLS = np.zeros_like(Tjul, dtype=int)

CLS = np.where(Tjul < 75, 1, CLS)
print(CLS)
CLS = np.where( (Tjul < 80) & (Tjul >= 75), 2, CLS)
print(CLS)
CLS = np.where(Tjul >= 80, 3, CLS)
print(CLS)
```
