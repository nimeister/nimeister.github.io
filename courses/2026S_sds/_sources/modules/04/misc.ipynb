```python
import numpy as np
```

```python
a = np.arange(6, dtype=np.float32).reshape(2, 3)

# Same shape & dtype as a → zeros
z1 = np.zeros_like(a)                 # shape (2,3), dtype float32
print(z1)
```

```python
row = np.arange(4)[None, :]           # shape (1, 4)
print(row)
col = np.arange(3)[:, None]           # shape (3, 1)
print(col)
row + col                  
```

```python
# 2D + 1D along last axis
A = np.ones((5, 3))                   # (5, 3)
print(A)
b = np.array([10, 20, 30])            # (3,)
print(b)
A + b                                 # result (5, 3)
```

```python
# Mismatched unless a size is 1:
X = np.ones((2, 3))
print(X)
Y = np.ones((4, 1))
print(Y)
# X + Y  → error (2 ≠ 4 and neither is 1 on the first axis)
```

```python
SEED = 123
np.random.seed(SEED)
Y, X = np.mgrid[0:8, 0:12]
#noise = np.random.normal(0, 0.05, size=(80,120))
D = 0.3*np.sin(X/15) + 0.25*np.cos(Y/21) 
print(D)

CLS = np.zeros_like(D, dtype=int)
print(CLS)
CLS = np.where(D >= 0.0, 1, CLS)
print(CLS)
CLS = np.where(D >= 0.3, 2, CLS)
print(CLS)
CLS = np.where(D >= 0.4, 3, CLS)
print(CLS)
D.shape, CLS.shape, np.unique(CLS)
```

```python
x = np.array([3, 1, 2, 3, 2, 2])
a = np.unique(x)
print(a)
```

```python
a = np.array([3, -1, 5, -2])

# Replace negatives with 0
b = np.where(a < 0, 0, a)              # → [3, 0, 5, 0]
print(b)
```

```python
# Pick elementwise from two arrays
b = np.array([10, 20, 30, 40])
np.where(a % 2 == 0, a, b)         # even → from a, odd → from b

# Multiple conditions (remember parentheses)
np.where((a < 0) | (a > 4), 99, a) # <0 or >4 → 99, else keep a
```
