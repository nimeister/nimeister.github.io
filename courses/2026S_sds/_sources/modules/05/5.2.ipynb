In this section, we will take a look at creating, renaming, and dropping columns; type conversion; and sorting &ndash; all of which make our analysis easier. We will be working with the 2019 Yellow Taxi Trip Data provided by NYC Open Data.

### Dropping columns
Let's start by dropping some columns which we won't be using.

```python
tree2005.columns
```

```python
columns2drop = ['cen_year', 'address', 'tree_loc', 'pit_type',
       'soil_lvl', 'status', 'spc_latin',  'vert_other',
       'vert_pgrd', 'vert_tgrd', 'vert_wall', 'horz_blck', 'horz_grate',
       'horz_plant', 'horz_other', 'sidw_crack', 'sidw_raise', 'wire_htap',
       'wire_prime', 'wire_2nd', 'wire_other', 'inf_canopy', 'inf_guard',
       'inf_wires', 'inf_paving', 'inf_outlet', 'inf_shoes', 'inf_lights',
       'inf_other', 'trunk_dmg', 'zipcode', 'zip_city', 'cb_num', 
        'cncldist', 'st_assem', 'st_senate', 'nta', 'nta_name',
       'boro_ct', 'state', 'latitude', 'longitude', 'x_sp', 'y_sp',
       'objectid_1', 'census tract', 'bin', 'bbl', 'Location 1']
tree2005 = tree2005.drop(columns=columns2drop)
```

### Groupbying and Sorting by values
`groupby` is a very useful pandas function to do many things. 
We can use the `sort_values()` method to sort based on any number of columns:

```python
# Find the most dominant species and tree counts
treecount2005 = tree2005.groupby('spc_common').size()
treecount2005_sorted = treecount2005.sort_values(ascending=False)
treecount2005_sorted[0:20]
```

```python
# Select big trees 
bigtree2005 = tree2005.loc[tree2005['tree_dbh'] > 40]
bigtree2005.shape
```

```python
# Find the most dominant species and tree counts in big trees
bigtreecount2005 = bigtree2005.groupby('spc_common').size()
bigtreecount2005_sorted = bigtreecount2005.sort_values(ascending=False)
bigtreecount2005_sorted[0:10]
```

## Working with the index

So far, we haven't really worked with the index because it's just been a row number; however, we can change the values we have in the index to access additional features of the pandas library.

### Setting and sorting the index

Currently, we have a RangeIndex, but we can switch to an index when calling `set_index()`:

```python
tree2005
```

```python
tree2005 = tree2005.set_index('borocode')
```

```python
tree2005
```

Since we have a sample of the full dataset, let's sort the index to order borocode:

```python
tree2005 = tree2005.sort_index()
tree2005
```

When not specifying a range, we use `loc[]`:

```python
tree2005.loc[5]
```

### Resetting the index

Sometimes we want to reset our index to row numbers and restore the columns with the `reset_index()` method:

```python
tree2005 = tree2005.reset_index()
tree2005.head()
```
