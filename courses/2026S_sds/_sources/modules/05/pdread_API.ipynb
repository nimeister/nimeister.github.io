## Getting data through an API

### API stands for Application Programming Interface.

It’s like a restaurant menu: you don’t go into the kitchen yourself, you tell the waiter (API) what you want, and the kitchen (server/database) sends back the dish (data).

In data terms: instead of downloading a giant file manually, you can “ask” the server for just the data you need, in the format you want.

### Getting Data Through an API

When you get data “through an API”: Your program sends a request (like typing a special URL, often with filters).
Example:

`https://data.cityofnewyork.us/resource/uvpi-gqnh.json?$limit=5`

→ asks NYC’s Open Data API for 5 rows of tree census data.

The API processes your request on the server.

It checks what you asked for (columns, filters, limits). 

It pulls that subset from the database. 

The API returns a response — usually in JSON (structured text) or XML. 

Your program can then parse this into a table or dataframe.

### Why Use an API?
1. Efficiency: Get only the rows/columns you need instead of a massive download.
2. Freshness: Data comes straight from the source, always up to date.
3. Automation: Your scripts can fetch data automatically every day/week.
4. Consistency: Same format and rules each time, so your code doesn’t break easily.


```python
import requests, pandas as pd

SERVICE = "https://data.cityofnewyork.us/api/odata/v4/uvpi-gqnh"
params = {
    "$top": 5,            # first 5 rows
    "$format": "json"
}
r = requests.get(f"{SERVICE}/rows", params=params)
r.raise_for_status()
df = pd.DataFrame(r.json()["value"])
df.head()
```
