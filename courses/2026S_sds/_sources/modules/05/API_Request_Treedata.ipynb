```python
import requests
import pandas as pd

api_url = "https://data.cityofnewyork.us/resource/uvpi-gqnh.json?$limit=50&$offset=150"  # Replace with your actual API URL
response = requests.get(api_url)

if response.status_code == 200:
    data = response.json()
    #print("Fetched data:", data)

tree2015_df = pd.DataFrame(data) # converts the returned data into a pd dataframe 

tree2015_df.tree_id
# tree2015_df.columns
# tree2015_df.index
```

 I am limiting the request to 50 items at a time using "li
 mit=50" and starting from page “4th page” of records (records 151 - 200) using "offset=150". I got this info from the nyc open data api window.
