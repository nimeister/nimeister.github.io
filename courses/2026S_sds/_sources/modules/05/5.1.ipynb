## 1. Pandas Series 
- A Pandas Series is a one-dimensional array of indexed data. The Pandas Series wraps both a sequence of values and a sequence of indices.
- The Pandas Series has an explicitly defined index associated with the values. The index can be integers or values of any desired type. 
- We can access the values and index attributes. 

````python
#Pandas series can be created from a list or array as follows.
data = pd.Series([0.25, 0.5, 0.75, 1.0])
print(data)
print(type(data))
````

````python
# Data can also be accessed by the associated index
data.values
````

````python
# Data can also be accessed by via the familiar Python square-bracket notation.
data[3]
````

````python
data[:]
````

````python
data.index
````

````python
# The index can be integers or values of any desired type.
# use strings as an index: 
data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=['a', 'b', 'c', 'd'])
print(data)
data['a']
````

````python
data['a']
````

## Extracting subsets

A crucial part of working with DataFrames is extracting subsets of the data: finding rows that meet a certain set of criteria, isolating columns/rows of interest, etc. After narrowing down our data, we are closer to discovering insights. This section will be the backbone of many analysis tasks.

#### Selecting columns

We can select columns as attributes if their names would be valid Python variables:

````python
tree2015.tree_id
````

You can select multiple columns at once this way:

````python
tree2015[['tree_id', 'spc_common']]
````

#### Selecting rows

````python
# In pandas, the [] (square bracket) syntax does not accept row and column selectors together like NumPy does.
tree2015[100:104, ['tree_id','spc_common']]
````

#### Indexing

Instead using `iloc[]` to select rows and columns by their position:

````python
tree2015.iloc[100:104, [0, 3, 4, 6]]
````

We use `loc[]` to select by name:

````python
tree2015.loc[100:104, 'tree_id':'tree_dbh']
````

````python
tree2015.loc[100:104, ['tree_id', 'tree_dbh']]
````

Please notice the difference between iloc and loc.

#### Filtering with Boolean masks

A **Boolean mask** is a array-like structure of Boolean values &ndash; it's a way to specify which rows/columns we want to select (`True`) and which we don't (`False`).

Here's an example of a Boolean mask for a subset of `red_maple` species with `tree_dbh` > 40 inches

````python
tree2015[(tree2015['tree_dbh'] > 40) & (tree2015.spc_common == 'red maple')]
````

**Important**: Take note of the syntax here. We surround each condition with parentheses, and we use bitwise operators (`&`, `|`, `~`) instead of logical operators (`and`, `or`, `not`).

Use logical operators (and, or, not) → for plain Booleans in conditions/flow control.

Use bitwise operators (&, |, ^, ~, etc.) → for binary math, masks, or array-wise logic.

With pandas/NumPy, you must use &, |, ~ for combining conditions, not and/or:

*Tip: Boolean masks can be used with `loc[]` and `iloc[]`.*

````python
tree2015[(tree2015['tree_dbh'] > 40) & (tree2015.spc_common == 'red maple')].iloc[0:10]
````

## Calculating summary statistics
We can extract some interesting insights from the `tree2015` data by calculating summary statistics.

#### What was the mass of the average tee_dbh

````python
tree2015['tree_dbh'].mean()
````

**Important**: The mean isn't always the best measure of central tendency. If there are outliers in the distribution, the mean will be skewed. Here, the mean is being pulled higher by some very heavy meteorites &ndash; the distribution is [right-skewed](https://www.analyticsvidhya.com/blog/2020/07/what-is-skewness-statistics/).

Taking a look at some quantiles at the extremes of the distribution shows that the mean is between the 95th and 99th percentile of the distribution, so it isn't a good measure of central tendency here:

````python
tree2015['tree_dbh'].quantile([0.01, 0.05, 0.5, 0.95, 0.99])
````

A better measure in this case is the median (50th percentile), since it is robust to outliers:

````python
tree2015['tree_dbh'].median()
````

#### What was the largest tree?

````python
tree2015['tree_dbh'].max()
````

````python
tree2015.loc[tree2015['tree_dbh'].idxmax()]
````

#### How many different types of tree species are in this dataset and what are they? 

````python
tree2015['spc_common'].nunique() # How many distinct values?
````

Some examples:

````python
tree2015['spc_common'].unique() # What are the distinct values?
````

#### Get some summary statistics on the data itself
We can get common summary statistics for all columns at once. By default, this will only be numeric columns, but here, we will summarize everything together:

````python
tree2015['tree_dbh'].describe(include='all')
````
