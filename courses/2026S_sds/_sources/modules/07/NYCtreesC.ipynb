## Carbon Storage, Tree growth and Caron Sequestration Rate for NYC Street Trees 

With NYC street tree species and dbh measurements, carbon storage for each tree can be calculated based on allometric equations. Half of he  above-ground biomass (AGB) is carbon stored by trees. AGB = wood density * woody volume. Wood volume can be estimated as a function of dbh (or dbh and tree height) using allometric equations. Urban tree allometric equations (woody volume and wood density) can be found [here](https://www.fs.usda.gov/rds/archive/Catalog/RDS-2016-0005).
