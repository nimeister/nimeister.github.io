## 2. Booleans
- There are two boolean-type objects: True and False; they belong to the built-in type bool.
- Three logic operators:
    1. and-`&` 
    2. or-'|`
    3. not 

```python
# Two types 
print(type(True))
print(type(False))
```

```python
print(True or False)
print(True and False)
print(not False)
```

```python
# You can use Operator symbols 
print(True & False)
print(True | False)
print( not True)
```

```python
print(10 > 9)
print(10 == 9)
print(10 < 9)
```

```python
a = 200
b = 33

if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
```

```python
# Most Values are True 
# Almost any value is evaluated to True if it has some sort of content.

# Any string is True, except empty strings.

# Any number is True, except 0.

# Any list, tuple, set, and dictionary are True, except empty ones.
print(bool("Hello"))
print(bool(15))
print(bool(False))
print(bool(None))
print(bool(0))
print(bool(""))
print(bool(()))
print(bool([]))
print(bool({}))
```

```python
print(bool(0))
bool(1)
```

## 3. NoneType 
- `'None` - `NoneType`,  nothing
- Python reserves `is` as an operation that checks if two objects are identical. This is different than ==, which checks if two objects are associated with the same value or state.
- 'None` is often used as a placeholder


```python
type(None)
```

```python
type(0)
```

```python
print(type(22) is None)
print(type(0) is None)
```

## 4. Strings

- can use single quotes or double or triple quotes
- `\n` is a single character, denoting a new line
- `\t` is a tab-character
- [A list of string functions](https://docs.python.org/3/library/stdtypes.html#string-methods)

```python
type(["hello",'World']) 
```

```python
print('Hello \n World')
print('hello \t World')
```

```python
# convert lower case of string
print('MyFunc'.lower())
```

```python
# replace string
print('MyFunc'.replace('Func','func'))
```
