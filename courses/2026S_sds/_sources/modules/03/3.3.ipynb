# Basics of Functions
A Python function is an object that encapsulates a segment of code. You can have inputs and outputs for each function. 
Calling the function will execute the encapsulated code and return an object. A function can be defined to accept arguments, 
which are objects to be passed to the encapsulated code.

```python

    <encapsulated code>
    return <object>
```

## The_return_Statement

The `return` statement is also reserved by Python. It denotes the end of a function; if reached, a return statement immediately concludes the execution of the function and returns the specified object. It can also return an empty statement or can be omitted. In both cases, it returns `None` object. 


````python
def evennum(x):
    return x % 2 == 0
evennum(3)
    
````

````python
# This function returns `None`
# an `empty` statement
def f():
    x=2
    print(x)
    return
f()
````

````python
# This function returns `None`
# return statement is omitted
def f():
    x=1
    print(x)
f()
````

## Arguments

Arguments in a Function are a sequence of comma-separated variable names. It can pass its argument in several ways: 

### Specifying Arguments by Position

The objects passed to a function will be assigned to its input variables based on their positions. 

````python
def bigger(a,b):
    if a > b: return a>b 
    else: return False
bigger(4,3)
````

````python
bigger(3,4)
````

````python
def is_bounded(x, lower, upper):
    return lower <=x <= upper
is_bounded(5,4,6)
````

### Specifying Arguments by Name 

You can provide explicit names when specifying the inputs to a function, in which case ordering does not matter. 

````python
is_bounded(x=4, lower=5, upper=7)
````

````python
#You can mix-and-match positional and named input by using position-based inputs first:
is_bounded(4,lower=5,upper=7)
````

````python
# Note that if you provide a named input, all the inputs 
# following it must also be named. Otherwise, you get an error:
is_bounded(4,lower=5, upper=7)
````

````python
is_bounded(2,4,upper=7)
````

### Default-Valued Arguments

You can specify default values for input arguments to a function. The default values are used if a user does not specify these inputs when calling the function. For example, by default, `y' is not a vowel. But if you want to include `y` as a vowel, we can use the default-valued argument:  

````python
def count_vowels(in_str, include_y=False):
    vowels = 'aeiouAEIOU'
    if include_y:
        vowels += 'yY'
    return sum(1 for char in in_str if char in vowels)
count_vowels('happy')
````

````python
# The default value can be overridden:
count_vowels('happy',include_y=True)
````

````python
# The default value can be overriden:
count_vowels('happy',True)
````

### Accommodating an Arbitrary Number of Positional Arguments 

An arbitrary number of positional arguments can be called by the syntax def f(*<var_name>). 
A function can accept any number of positional arguments, not a fixed count.
This syntax can be combined with positional arguments and default arguments. 
Any variables specified after a packed variable must be called by name.
  
Python cannot foresee how many arguments will be passed to the function, all of the objects that are passed to it will be packed into a tuple, which is then assigned to the variable args:
```python
def f(*args):
    #  All keyword arguments passed to `f` will be "packed" into a
    #  dictionary that is assigned to the variable `args`.
    # `f()`  will assign `args = {}` (an empty dictionary)
    # `f(x=1, y=2, ...)` will assign `args = {"x":1, "y":2, ...}`
    return args

f() - pass zero argument to `f`
{}
f(x) - pass one argument to `f`
{'x':1}
f(x=(0,1), val=True) - pass two arguments to `f` 
{'x': (0,1), 'val': True}
```

This syntax can be combined with positional arguments and default arguments. Any variables specified after a packed variable must be called by name:

````python
def number_of_args(*args):
    return len(args)

print(number_of_args(2,3,None))
````

This syntax can be combined with positional arguments and default arguments. Any variables specified after a packed variable must be called by name:

````python
def f(x, *seq, y):
    print("x is: ", x)
    print("seq is: ", seq)
    print("y is: ", y)
    return None
f(1,2,4, y=5)
````

````python
f('cat',y='dog')
````

* indicates the packing of an arbitrary number of arguments into a tuple, when used in the signature of a function definition. Simultaneously, * signals the unpacking of an iterable to pass each of its members as a positional argument to a function, when used in the context of calling a function:

````python
def f(x,y,z):
    return x+y+z
f(2,4,6)
````

````python
# you can implement in this way: 
f(*[2,4,6])
# * means unpack the content of [2,4,6] and pass them to x,y,z, respectively.
````

### Accommodating an Arbitrary Number of Keyword Arguments 

You can define a function that is able to accept an arbitrary number of keyword arguments, using the syntax: def f(**<var_name>)

Note that a single asterisk, *, was used to denote an arbitrary number of positional arguments, whereas ** signals the acceptance of an arbitrary number of keyword arguments.

 ** indicates the packing of an arbitrary number of keyword arguments into a dictionary, when used in the signature of a function definition. Simultaneously, ** signals the unpacking of a dictionary to pass each of its key-value pairs as a keyword argument to a function, when used in the context of calling a function:

Python cannot foresee how many arguments will be passed to f, all of the keyword arguments that are passed to it will be packed into a dictionary, where a given keyword is set as a key (cast as a string) that maps to the corresponding value. This dictionary is then assigned to the variable args. Dictionaries will be discussed in detail in a later section.

````python
def f(**kargs):
    return(print(kargs))
f(a=1,b=2,c=3)
    
````

````python
def f(x, y=2, **kargs):
    print(x)
    print(kargs)
    return None
    
f(3, y=True, k=[3,4])
````

````python
# accepting arbitary positional and keyword arguments
def f(*x, **y):
    # all positional arguments get into the tuple `x`
    # all the keyword argumments get into the dictionary `y`
    print(x)
    print(y)
    return None
f(1,3, hi=1,bye=2)
````

## Geospatial Example: Haversine Function

Let's apply these concepts to a geospatial problem. The [Haversine formula](https://en.wikipedia.org/wiki/Haversine_formula) calculates the distance between two points on the Earth’s surface.

![](https://upload.wikimedia.org/wikipedia/commons/c/cb/Illustration_of_great-circle_distance.svg)

````python
import math as m
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius in kilometers
    dlat = m.radians(lat2 - lat1)
    dlon = m.radians(lon2 - lon1)
    a = (
        m.sin(dlat / 2) ** 2
        + m.cos(m.radians(lat1)) * m.cos(m.radians(lat2)) * m.sin(dlon / 2) ** 2
    )
    c = 2 * m.atan2(m.sqrt(a), m.sqrt(1 - a))
    distance = R * c
    return distance


# Example usage
distance = haversine(35.6895, 139.6917, 34.0522, -118.2437)
print(f"Distance: {distance:.2f} km")
````
