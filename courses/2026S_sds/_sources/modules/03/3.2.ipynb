### Python if Statements

This can be very useful for checking whether a certain condition exists before the code begins to execute, as you may want to only execute certain code lines when certain conditions are met.   

For example, conditional statements can be used to check that a certain variable or file exists before code is executed, or to execute more code if some criteria are met, such as a calculation resulting in a specific value. 

A conditional statement's structure is defined through `if` and `else`. 

```python
if condition1:
    print('condition1 is true')
elif condition2:
    print('condition2 is true')
else:
    print("conditions not true")
```

When the condition following the `if` condition1 is met, then a certain code will execute, otherwise if condition2, excute certain code.  When both conditions are not met, then the code following the `else` will execute.


````python
# create a variable x with the value of 0
x = 30
# compare the value of x to 0 and 10 and print a message
if x <= 0:   
    print('x has a value of', x, 'which is <= 0.')
elif x < 10:
   print('x has a value of', x, 'which is between (0 10).') 
elif x <20:
    print('x has a value of',x, 'which is between [10,20)')
else:
    print('x has a value of', x, 'which is not less than 20.') 
          
x= x+1
````

You can also write conditional statements to identify keywords within a text string, such as the name of a variable. 

````python
species = ['London Planetree', 'HoneyLocust', 'Callery Pear', 'Oak','Norway Maple' ]
# check if the text string "Oak" in contained within the species list
if 'Ginkgo' in species:
    print('Ginkgo is one of the most NYC popular tree species')
else:
    print('Ginkgo is not one of the most NYC popular tree species')  
````

````python
coordinates = [
    (35.6895, 139.6917),
    (34.0522, -118.2437),
    (51.5074, -0.1278),
]  # List of tuples representing coordinates

for lat, lon in coordinates:
    if lat > 0:
        print(f"{lat} is in the Northern Hemisphere")
    elif lat < 0:
        print(f"{lat} is in the Southern Hemisphere")
    else:
        print(f"{lat} is near the equator")
````
