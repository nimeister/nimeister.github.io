## 1. Lists
Python knows a number of compound data types, used to group together other values. The most versatile is the list, 
which can be written as a list of comma-separated values (items) between square brackets. 
Lists might contain items of different types, but usually the items all have the same type.
- A sequence of objects
- is created using sqare-brackets `[]`, its items (no need to the same type) are seperated by commas. 

```python
[3.5, 'a']
```

```python
isinstance([1,'a'], list)
```

```python
# [] - an empty list
[]
```

```python

# list can be a set of numbers
dbh = [2, 15, 20,25] # in inch
print(dbh[3])
print(dbh[:])
print(dbh[0])

# list can be a set of strings
species = ['London Planetree', 'HoneyLocust', 'Callery Pear', 'Oaks','Norway Maple' ]
print(species[-1])
           
# List can be a mix of numbers and strings 
tree =  ['London Planetree', 26, 'HoneyLocust', 20, 'Callery Pear', 5, 'Oaks', 30, 'Norway Maple', 27] 
print(tree[-2:])
print(tree[:-2])
print(tree[:3])
```

### More on lists 
The list data type has some methods. Here are a few methods for listing objects:

```python
# `list` forms a list of content of a sequence
list('apple')
```

```python
species = ['London Planetree', 'HoneyLocust', 'Callery Pear', 'Oak','Norway Maple' ]
print(species)

# list.append(x)
# Add an item to the end of the list. Equivalent to a[len(a):] = [x].
species.append('Ginkgo')
print('append Ginkgo at the end:', species)
```

```python
# list.insert(i, x)
# Insert an item at a given position. The first argument is the index of the element
# before which to insert, so a.insert(0, x) inserts at the front of the list, and 
# a.insert(len(a), x) is equivalent to a.append(x).
species.insert(0,'Ginkgo')
print('after add Ginkgo:', species)
```

```python
# list.remove(x)
# Remove the first item from the list whose value is equal to x. It raises a ValueError if there is no such item.
species.remove('Ginkgo')
print('after remove Ginkgo:', species)

species.append('Noway Spruce')
print('append species', species[-1])

# list.index(x[, start[, end]])
# Return zero-based index in the list of the first item whose value is equal to x. Raises a ValueError if there is no such item.
print('Index of Norway Maple', species.index('Norway Maple'))
```

```python
# list.count(x)
# eturn the number of times x appears in the list.
print('Count of Oak trees:',species.count('Oak'))
print(species.count('Ginkgo'))

print(species)
# list.sort(*, key=None, reverse=False)
# Sort the items of the list in place 
species.sort()

print('sorted species:',species )
species.sort(reverse=True)
print('sorted species reversed:', species)
```

```python
# list.reverse()
# Reverse the elements of the list in place.
species.reverse()
print('species_reverse', species)
```

```python
# list.copy()
# Return a shallow copy of the list. Equivalent to a[:].
sp = species.copy()
print('copied species:', sp)

# list.clear()
# remove all items from the list. Equivalent to del a[:].
sp.clear()
print('copies species after clear:', sp)
```

## 2. Tuples
A tuple is a collection that is ordered and unchangeable. A tuple consists of a number of values separated by commas with round brackets. Tuples are ***ordered, unchangeable, allow duplicates***

```python
t = (12345, 54321, 'hello!',12345)
print(t)

# Join two tuples
t1 = (2, 4, 5)
print(t+t1)


u = t, (1,2,3)
print(u)

# Access Tuple Items
print(u[1])
print(u[-2])
```

```python
# One item tuple, add the comma in the end
thistuple = ("apple",)
print(type(thistuple))

#NOT a tuple
thistuple = ("apple")
print(type(thistuple))

# Multiply tuples 
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2

print(mytuple)
```

## 3. Sets
Sets are used to store multiple items in a single variable. A set is a collection which is both unordered and unindexed. Sets are written with curly brackets.


```python
thisset = {'apple', 'banana', 'Cherry'}
print(thisset)

#Loop through the set
for x in thisset:
  print(x)

# Check if 'banana' is in the set
print('banana' in thisset)
print('pear' in thisset)

# Duplicates not allowed
thisset = {'apple', 'banana', 'cherry', 'apple'}
print(thisset)

# get the length 
print(len(thisset))

# sets can be any types 
set1 = {'abc', 34, True, 40, 'male', 34}
print(len(set1))
```

```python
# Loop through the set, and print the values:
for x in thisset:
  print(x)

# add() and remove() items

thisset.add('orange')
print(thisset)

thisset.remove('apple')
print(thisset)
```

```python
# join sets using union() or update()
set1 = {'apple', 'banana', 'cherry', 'apple'}
print(set1)
set2 = {'orange', 'kiwi','apple'}
print(set2)
set3 = set1.union(set2)
print(set1)
print(set3)
set1.update(set2)
print(set1)
```

```python
# Keep ONLY the Duplicates using the intersection_update() method 
set1 = {'apple', 'banana', 'cherry', 'apple'}
print(set1)
set2 = {'orange', 'kiwi','apple'}
print(set2)

set3 = set1.intersection(set2)
print(set3)

set1.intersection_update(set2)
print(set1)
```

```python
# Keep All, But NOT the Duplicates The symmetric_difference_update() method will keep only the elements
# that are NOT present in both sets.
set1 = {'apple', 'banana', 'cherry', 'apple'}
print(set1)
set2 = {'orange', 'kiwi','apple'}
print(set2)
set1.symmetric_difference_update(set2)
print(set1)
```

## 4. Dictionaries
Dictionaries are used to store data values in key:value pairs.
A dictionary is a collection which is ordered*, changeable and does not allow duplicates.

Dictionaries are collections of key-value pairs, where each key is unique. Dictionaries are extremely useful for storing data that is associated with specific identifiers, such as attribute data for geographic features.

For example, you can use a dictionary to store attributes of a geospatial feature, such as a city.

```python
car = {
  'brand': 'Ford',
  'model': 'Mustang',
  'year': 1964
}
print(car)
```

```python
# data type
print(type(car))

# add items
car['color'] = 'blue'
print(car)

# Get the value of the "model" key:
x = car['model']
print(x)

# or use get()
x = car.get('model')
```

```python
# The keys() method will return a list of all the keys in the dictionary.
x = car.keys()
print(x)

# Get a list of the values:
x = car.values()
print(x)
```

```python
# change values
x = car.values()
print(x) #before the change
car["color"] = "red"
print(x) #after the change
```

```python
city_attributes = {
    "name": "Tokyo",
    "population": 13929286,
    "coordinates": (35.6895, 139.6917),
}  # Dictionary storing attributes of a city
```

You can access the values associated with specific keys:

```python
city_name = city_attributes["name"]
city_population = city_attributes["population"]
print('City:', city_name, 'Population:', city_population)
# you can also use formated string: put f in front of the string, then use {} to embed variables.
# print(f'City: {city_name} Population: {city_population}') 
```

You can also add or update key-value pairs in a dictionary:

```python
city_attributes["area_km2"] = 2191  # Adding the area of the city in square kilometers
print("Updated city attributes:", city_attributes)
```

# Working-with-Sequences

## Checking if an object is in a sequence: `obj in seq`

```python
seq = [3,5,6,7, 9, 5]
4 in seq
```

## Concatenating sequences: `seq1 +seq2`


```python
seq1=[1,2,3]
seq2=[3,4,5]
seq = seq1+seq2
print(seq)
```

## Repeated concatenation of sequences: `n*seq1` or `seq1*n`

```python
seq = [3,5,6,7, 9, 5]
2*seq
```

## Asking for the length of a sequence

```python
seq = [3,5,6,7, 9, 5]
len(seq)
```

## Getting the index of the first occurence of the x in a sequence: `seq.index(x)

```python
seq = [3,5,6,7, 9, 5]
seq.index(3)
```

## Counting the number of occurences of `x` in a sequence: `seq.count(x)

```python
seq = [3,5,6,7, 9, 5]
print(seq.count(3))
```

## Slicing
- Retrieve a subsequence of items, based on the indexing scheme
- seq[start:end:step]: start -the index where the slice starts (inclusive), end: the index where the slice ends (exclusive), step: step size, can be negative. 

```python
seq = [3,5,6,7, 9, 5]
print(seq[2:5:2])
print(seq[5:2:-2])
```

```python
# start 0, end last, step 1
print(seq[:])

# start 0, end last, step 2
print(seq[::2])
```
