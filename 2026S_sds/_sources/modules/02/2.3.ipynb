# Variables

- Within a programming workflow, variables store data (i.e., information) that you want to re-use in your code (e.g., a single value, a list of values, a filename, or a path to a directory). 

- In `Python`, variables can be created without defining the data type they will hold (e.g., whole numbers, text strings). 

- You can create a variable to store a single numeric value by defining a name and assigning a value to that name using the equal (`=`) operator (e.g., `variablename = 1.5`). 

- If the value is a text string, you need to add parentheses (`""`) around the value, so that `Python` knows it is a text string (e.g., `variablename = "text"`). Strings in Python are surrounded by either single quotation marks or double quotation marks. 'hello' is the same as "hello".

```python
# varibles can be numbers or strings. 
# You can display a string literal with the print() function. 
precip = 4.83
print(precip)

# define string using ' ' or " "
city = 'New York'
print(city)

print(type(precip))
print(type(city))
```

# Naming
A variable can have a short name (like x and y) or a more descriptive name (age, carname, total_volume). Rules for Python variables:
- A variable name must start with a letter or the underscore character
- A variable name cannot start with a numberic value
- A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
- Variable names are case-sensitive (age, Age and AGE are three different variables)
- They also cannot conflict with character sequences reserved by the Python language. As such, the following cannot be used as variable names:
   - `for`, `while`, `break`, `pass`, `continue`
    - `in`, `is`, `not`
    - `if`, `else`, `elif`
    - `def`, `class`, `return`, `yield`, `raises`
    - `import`, `from`, `as`, `with`
    - `try`, `except`, `finally`
    - `all`,`any`

```python
A = 3
print(a)
```

```python
# varible names are case sensitive
a = 3
A ='New York'
# A won't overwrite a
print(a)
print(A)
```

```python
# If you want to specify the data type of a variable, this can be done with casting.
x = str(3)    # x will be '3'
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0
```

```python
# Multi Words Variable Names
# Variable names with more than one word can be difficult to read.

# There are several techniques you can use to make them more readable:

# Camel Case
# Each word, except the first, starts with a capital letter:
myVariableName = 'Ted'

# Pascal Case
# Each word starts with a capital letter:
MyVariableName = 'Ted'

# Snake Case
# Each word is separated by an underscore character:

my_variable_name = 'Annie'
```

# Assignment
In Python, the = symbol represents the “assignment” operator. The variable goes to the left 
of =, and the object that is being assigned to the variable goes to the right:

```python
a = 10
```

```python
# Many Values to Multiple Variables
# Python allows you to assign values to multiple variables in one line
a, b, c = 1, 2, 3
print(a)
print(b)
print(c)

# One Value to Multiple Variables
# And you can assign the same value to multiple variables in one line:
x = y= z = 'Apple'
print('x = ', x)
print('y = ', y)
print('z = ', z)
```

# Mutable-Immutable Objects
- A mutable object can have its state changed. For instance, a list is an example of a mutable object. Once formed, we can update the contents of a list - replacing, adding to, and removing its elements. Some mutable objects:
    - lists
    - dictionaries
    - sets
    - NumPy arrays
      
- An immutable object cannot have its state changed. For instance, a tuple is an immutable object. Once formed, there is no mechanism by which one can change the state of a tuple, and any code that appears to be updating a tuple is in fact creating an entirely new tuple. Some immutable objects:
    - numbers (integers, floating-point numbers, complex numbers)
    - strings
    - tuples
    - booleans
    - “frozen”-sets


```python
# List is mutable
x = [1, 2, 3]
print(x)
x[1] = 5
print(x)
```

```python
# Tuple is immutable
x = (1,2,3)
print(x)
x[1]= 4
print(x)
```
